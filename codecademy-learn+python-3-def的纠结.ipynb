```python
def countdown(n):
    if n<=0:
        print("Blastoff!")
    else:
        print(n)
    countdown(n-1)
    return
countdown(3)#最后一个就是赋值n=3，也是输出。如果没有这一个，前面类似设计公式，没有输出或调用，computer木有反应
```
